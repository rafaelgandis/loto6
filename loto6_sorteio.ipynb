```python
import xml.etree.ElementTree as ET
from collections import Counter
import random

def ler_sorteios_xmls(nome_arquivo):
    tree = ET.parse(nome_arquivo)
    root = tree.getroot()

    sorteios = []

    for linha in root.findall('linha'):
        dezenas = [int(dezena.text) for dezena in linha.findall('dezena')]
        sorteios.append(dezenas)

    return sorteios

def gerar_sorteios(sorteios, num_sorteios):
    frequencias = Counter([dezena for sorteio in sorteios for dezena in sorteio])
    dezenas_ordenadas = sorted(frequencias, key=lambda x: frequencias[x], reverse=True)

    sorteios_gerados = []

    while len(sorteios_gerados) < num_sorteios:
        sorteio = random.sample(dezenas_ordenadas, 6)
        if sorteio not in sorteios_gerados:
            sorteios_gerados.append(sorteio)

    return sorteios_gerados

# Ler o arquivo XMLS com os sorteios
nome_arquivo = '/content/SORTEIOS_LOTO6.xlsx'
sorteios = ler_sorteios_xmls(nome_arquivo)

# Gerar 15 sorteios com base nas probabilidades
num_sorteios_gerados = 15
sorteios_gerados = gerar_sorteios(sorteios, num_sorteios_gerados)

# Exibir os sorteios gerados
for i, sorteio in enumerate(sorteios_gerados):
    print(f"Sorteio {i+1}: {sorteio}")
```

```python
import pandas as pd
import random

def ler_sorteios_xlsx(nome_arquivo):
    df = pd.read_excel(nome_arquivo)
    sorteios = df.values.tolist()
    return sorteios

def gerar_sorteios(sorteios, num_sorteios):
    dezenas = [dezena for sorteio in sorteios for dezena in sorteio]
    frequencias = pd.Series(dezenas).value_counts().sort_values(ascending=False)
    dezenas_ordenadas = list(frequencias.index)

    sorteios_gerados = []

    while len(sorteios_gerados) < num_sorteios:
        sorteio = random.sample(dezenas_ordenadas, 6)
        if sorteio not in sorteios_gerados:
            sorteios_gerados.append(sorteio)

    return sorteios_gerados


nome_arquivo = '/content/SORTEIOS_LOTO6.xlsx'
sorteios = ler_sorteios_xlsx(nome_arquivo)

# Gerar 15 sorteios com base nas probabilidades
num_sorteios_gerados = 20
sorteios_gerados = gerar_sorteios(sorteios, num_sorteios_gerados)


for i, sorteio in enumerate(sorteios_gerados):
    print(f"Sorteio {i+1}: {sorteio}")
```
